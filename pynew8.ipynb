```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/placement - placement.csv')
df.head()
```

```python
x=df.drop(columns = ['placed'])      #input data
y=df['placed']                       #target data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
np.round(x_train.describe(),2)   #we used np.round to round off the values
```

```python
x_train.describe()
```

```python
#mean = average

#Normal Distribution ---->central tendency(mean,median)---->left and right size equal data points.
#Standard deviation ----> Next value gap from central tendency
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)   #fit means learn the parameters and transform means apply on that data
```

```python
x_train_sc
```

```python
np.round(x_train_sc.describe(),2)
```

```python
type(x_train_sc)
```

```python
x_train_new=pd.DataFrame(x_train_sc, columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
