ANN(Artificial Neural Network)

#Steps of working with ANN

1. Data Import------->Data cleaning and preprocessing

2. Create architecture ----> Model(input_layer , Hidden_layer , output_layer)

3. Model_compile(optimizer , loss , metrics)

4. Model.fit(x_train , y_train)

5. Model total_loss , Total_accuracy

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Attrition - Attrition.csv')
df.head(2)
```

```python
import tensorflow as tf    ##deep learning library
from sklearn.model_selection import train_test_split     ##data dividation
from sklearn.preprocessing import LabelEncoder , StandardScaler     ##label for encoding and standard scalar for normalization
from sklearn.metrics import accuracy_score     ##performance evaluation metrics
```

```python
df = df.dropna()
```

```python
x = df.drop(columns=['Attrition'])
y = df['Attrition']
```

```python
x.info()
```

```python
categorical_features = ['BusinessTravel',
                        'Department',
                        'EducationField',
                        'Gender',
                        'JobRole',
                        'MaritalStatus',
                        'Over18',
                        'OverTime']
```

```python
x_encoded = pd.get_dummies(x , columns=categorical_features , drop_first=True)
```

```python
x_encoded = x_encoded.astype(int)
```

```python
x_encoded
```

```python
#encode target data
lb=LabelEncoder()
```

```python
y_encoded=lb.fit_transform(y)
y_encoded = pd.DataFrame(y_encoded)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_encoded , y_encoded , test_size=0.2 , random_state=42)
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
#Architecture of our model
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=128 , activation='relu' , input_dim=x_train_scaled.shape[1]),    #input layer
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=64 , activation='relu'),       #hidden layer
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=1 , activation='sigmoid')     #output layer
])
```

```python
#Now compile the model
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
metrics = ['accuracy']
```

```python
learning_rate = 0.001
momentum = 0.9
optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate , momentum=momentum , nesterov=True)
```

```python
model.compile(optimizer=optimizer , loss=loss_fn , metrics=metrics)
```

```python
#Train the model
```

```python
model.fit(x_train_scaled , y_train , epochs=50 , batch_size=8 , validation_split=0.1)
```

```python
loss , accuracy = model.evaluate(x_test_scaled , y_test)
```

```python
print("Total Accuracy : " , accuracy)
```
